```python
# risky payers/defaulters
risky_defaulter_file = "/home/prof/kaggle/unbankedcredit/risky_defaulters.csv"
risky_defaulters = pd.read_csv(risky_defaulter_file)
risky_defaulters.describe()
```

```python
# risky payers/defaulters
risky_payers_file = "/home/prof/kaggle/unbankedcredit/risky_payers.csv"
risky_payers = pd.read_csv(risky_payers_file)
risky_payers.describe()
```

```python
risky_d_merge = pd.merge(app_bur,risky_defaulters, on='SK_ID_CURR', how='inner')
d_averages = risky_d_merge.groupby(by='SK_ID_CURR').mean()
d_sums = risky_d_merge.groupby(by='SK_ID_CURR').sum()
```

```python
risky_p_merge = pd.merge(app_bur,risky_payers, on='SK_ID_CURR', how='inner')
r_averages = risky_p_merge.groupby(by='SK_ID_CURR').mean()
r_sums = risky_p_merge.groupby(by='SK_ID_CURR').sum()
```
