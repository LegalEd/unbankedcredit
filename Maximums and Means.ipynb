```python
def median_max(file):
    """function to return a dataframe of the median and max results in input csv file"""
    df = pd.read_csv(file, low_memory=True)
    print(df.info)
    df = df.apply(pd.to_numeric,errors="coerce", downcast="integer")
    df = df.apply(pd.to_numeric,errors="coerce", downcast="float")
    median = df.groupby('SK_ID_CURR').median()
    maximum = df.groupby('SK_ID_CURR').max()
    means_max = pd.merge(means, median, on="SK_ID_CURR", suffixes=("_MEDIAN","_MAX"))
    return means_max
    
```

```python
bureau = median_max("/home/prof/kaggle/unbankedcredit/kaggle/all/bureau.csv") 
```

```python
credit_card_balance = median_max("/home/prof/kaggle/unbankedcredit/kaggle/all/credit_card_balance.csv")
```

```python
payments = median_max("/home/prof/kaggle/unbankedcredit/kaggle/all/installments_payments.csv")
```

```python
pos = median_max("/home/prof/kaggle/unbankedcredit/kaggle/all/POS_CASH_balance.csv")
```

```python
previous = median_max("/home/prof/kaggle/unbankedcredit/kaggle/all/previous_application.csv")
```

```python
def merger(df1, df2):
    df = pd.merge(df1,df2,on="SK_ID_CURR", how="left")
    return df
```

```python
m1 = merger(bureau,credit_card_balance )
m1plus = merger(m1,payments)
m1plus2 = merger(m1plus,pos)
final = merger(m1plus2, previous)
```

Drop columns with only NaN

```python
final.dropna(axis=1, how='all', inplace=True)
```

```python
final_median = final.filter(regex="MEDIAN")
```

```python
final_max = final.filter(regex="MAX")
```

```python
combined = pd.merge(final_median, final_max, on="SK_ID_CURR")
```
