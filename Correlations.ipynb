```python
train_file = "/home/prof/kaggle/unbankedcredit/train_enhanced_loan_history_cc_pos_previous.csv"
train_pd = pd.read_csv(train_file, low_memory=True)
train_pd.info()

```

Downcasting to reduce memory overhead

```python
train_pd = train_pd.apply(pd.to_numeric,errors="coerce", downcast="integer")
```

```python
test_pd = test_pd.apply(pd.to_numeric,errors="coerce", downcast="integer")
```

```python
pd.options.display.max_rows = 100
# Find correlations with the target and sort
correlations = train_pd.corr()['TARGET'].sort_values(ascending=False)

```

```python
# # Display correlations
# print('Most Positive Correlations:\n', correlations.tail(100))
# correlations
pos_list = [x[0] for x in correlations.tail(100).items()]
print(pos_list)
```

```python
print('\nMost Negative Correlations:\n', correlations.head(100))
```

```python
neg_list = [x[0] for x in correlations.head(100).items()]
```

```python
neg_list
```

```python
all_list = pos_list+neg_list
```

```python
all_list.append('SK_ID_CURR')
```

```python
corr_pd = pd.read_csv(train_file, usecols=all_list,)
corr_pd.info()
```

1 hot encoding object

```python
corr_pd = pd.get_dummies(corr_pd)
```

```python
corr_pd.head()
```

```python
corr_pd.info()
```

Exaggreate data by multiplying columns by themselves

```python
exagg_id_target = corr_pd[['SK_ID_CURR', 'TARGET']]
```

```python
corr_pd.drop(columns=['SK_ID_CURR', 'TARGET'], inplace=True)
exaggerated = corr_pd.multiply(corr_pd)
```

```python
exaggerated.head()
```

```python
exaggerated = pd.merge(exagg_id_target,exaggerated, left_index=True, right_index=True)
```

```python
exaggerated.head(100)
```

```python
exaggerated.to_csv("train_exaggreated_data.csv", index=False)
```
