Mean and sums for each loan ID

```python
cc['CC_COUNT'] = cc.groupby('SK_ID_CURR')['SK_ID_CURR'].transform('count')
```

```python
cc_counts = cc[['SK_ID_CURR', 'CC_COUNT']] # make separate df with just loan and counts to add back later
dropped = cc.drop(columns="CC_COUNT")

means = dropped.groupby('SK_ID_CURR').mean()
```

```python
sums = dropped.groupby('SK_ID_CURR').sum()
```

```python
means_sums = pd.merge(means,sums, on="SK_ID_CURR", how="inner", suffixes=("_MEANS","_SUMS"))
```

```python
means_sums.head()
```

```python
means_sums_counts = pd.merge(means_sums, cc_counts, on=["SK_ID_CURR"], how="left")
```
